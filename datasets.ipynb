```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import numpy as np
import matplotlib.pyplot as plt
from utils import version_check, load_dataset_cat, load_dataset_sklearn_moon, load_dataset_planar, load_dataset_sklearn_circle
from utils import load_dataset_2D

%matplotlib inline
```

## 1. Cat dataset

```python
train_X, train_Y, test_X, test_Y, classes = load_dataset_cat()
train_X.shape, train_Y.shape, test_X.shape, test_Y.shape, list(classes)
```

```python
# visualize random five images with labels in train set
np.random.seed(0)
random_idx = np.random.randint(0, train_X.shape[0], size=6)
print(random_idx)

plt.figure()
subplot = 1
plt.subplots_adjust(hspace=0.4)
for i in random_idx:
    img = train_X[i]
    label = np.squeeze(train_Y[:, i]) # [label] --> label 
    classes[label]
    plt.subplot(2, 3, subplot)
    subplot += 1
    plt.imshow(img)
    plt.title(classes[label].decode('utf-8'))
plt.show()
```

```python
# visualize random five images with labels in test set
np.random.seed(2)
random_idx = np.random.randint(0, test_X.shape[0], size=6)
print(random_idx)

plt.figure()
subplot = 1
plt.subplots_adjust(hspace=0.4)
for i in random_idx:
    img = test_X[i]
    label = np.squeeze(test_Y[:, i]) # [label] --> label 
    classes[label]
    plt.subplot(2, 3, subplot)
    subplot += 1
    plt.imshow(img)
    plt.title(classes[label].decode('utf-8'))
plt.show()
```
