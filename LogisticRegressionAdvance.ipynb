## 2. Regularizer

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import numpy as np
import matplotlib.pyplot as plt
from utils import version_check, load_dataset_cat, load_dataset_sklearn_moon, load_dataset_planar, load_dataset_sklearn_circle
from utils import plot_decision_boundary, print_mislabeled_images, load_dataset_2D
from LogisticRegressionRegularizer import LogisticRegression
%matplotlib inline
```

### 2D dataset

```python
train_X, train_Y, test_X, test_Y = load_dataset_2D(True)
```

#### L2 regularization

```python
layer_dims = [train_X.shape[0], 20, 3, 1]
clf = LogisticRegression(layer_dims, lr=0.3, n_epochs=30000, step_size=1000, initializer='xavier', lambd=0.7)
```

```python
# training
clf.fit(train_X, train_Y)
clf.plot_cost()

# train and test set evaluation
clf.evaluate(train_X, train_Y)
clf.evaluate(test_X, test_Y)
```

```python
plt.title("Model with L2-regularization")
axes = plt.gca()
axes.set_xlim([-0.75, 0.40])
axes.set_ylim([-0.75, 0.65])
plot_decision_boundary(lambda x: clf.predict(x.T), train_X, train_Y)
```

#### Dropout

```python
# model creation
clf = LogisticRegression(layer_dims, lr=0.3, n_epochs=30000, step_size=1000, initializer='xavier', keep_prob=0.86)
# training
clf.fit(train_X, train_Y)
clf.plot_cost()

# train and test set evaluation
clf.evaluate(train_X, train_Y)
clf.evaluate(test_X, test_Y)
```

```python
plt.title("Model with Dropout({})".format(clf.keep_prob))
axes = plt.gca()
axes.set_xlim([-0.75, 0.40])
axes.set_ylim([-0.75, 0.65])
plot_decision_boundary(lambda x: clf.predict(x.T), train_X, train_Y)
plt.show()
```

## 3. Optimizer

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import numpy as np
import matplotlib.pyplot as plt
from utils import version_check, load_dataset_cat, load_dataset_sklearn_moon, load_dataset_planar, load_dataset_sklearn_circle
from utils import plot_decision_boundary, print_mislabeled_images, load_dataset_2D
from LogisticRegressionOptimizer import LogisticRegression
%matplotlib inline
```

### Moon dataset

```python
train_X, train_Y = load_dataset_sklearn_moon(True)
```

#### Minibatch Gradient Descent

```python
layer_dims = [train_X.shape[0], 5, 2, 1]
clf = LogisticRegression(layer_dims, optimizer='gd', n_epochs=10000, step_size=500, lr=0.0007)
clf.fit(train_X, train_Y)
clf.plot_cost()
clf.evaluate(train_X, train_Y)
plot_decision_boundary(lambda x: clf.predict(x.T), train_X, train_Y)
```

#### Momentum

```python
layer_dims = [train_X.shape[0], 5, 2, 1]
clf = LogisticRegression(layer_dims, optimizer='momentum', n_epochs=10000, step_size=500, lr=0.0007, beta=0.9)
clf.fit(train_X, train_Y)
clf.plot_cost()
clf.evaluate(train_X, train_Y)
plot_decision_boundary(lambda x: clf.predict(x.T), train_X, train_Y)
```

#### Adam

```python
layer_dims = [train_X.shape[0], 5, 2, 1]
clf = LogisticRegression(layer_dims, optimizer='adam', n_epochs=10000, step_size=500, lr=0.0007)
clf.fit(train_X, train_Y)
clf.plot_cost()
clf.evaluate(train_X, train_Y)
plot_decision_boundary(lambda x: clf.predict(x.T), train_X, train_Y)
```

#### Adam with Dropout

```python
layer_dims = [train_X.shape[0], 5, 2, 1]
clf = LogisticRegression(layer_dims, optimizer='adam', n_epochs=10000, step_size=500, lr=0.0007, keep_prob=0.9)
clf.fit(train_X, train_Y)
clf.plot_cost()
clf.evaluate(train_X, train_Y)
plot_decision_boundary(lambda x: clf.predict(x.T), train_X, train_Y)
```
